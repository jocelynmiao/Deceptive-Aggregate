```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import zarr
import gcsfs
```

```python
df = pd.read_csv('https://storage.googleapis.com/cmip6/cmip6-zarr-consolidated-stores.csv')
df_ta = df.query("activity_id=='CMIP' & table_id == 'Amon' & variable_id == 'tas' & experiment_id == 'historical'")
df_ta_ncar = df_ta.query('institution_id == "NCAR"')
gcs = gcsfs.GCSFileSystem(token='anon')

#example stuff from their demo nb
zstore = df_ta_ncar.loc[59875].zstore
mapper = gcs.get_mapper(zstore)
ds = xr.open_zarr(mapper, consolidated=True)
```

```python
import geopandas as gpd
import numpy as np
import regionmask
import xarray as xr

# 1. Standardize longitudes to [-180, 180] and sort on the full dataset
ds = ds.assign_coords(lon=(((ds.lon + 180) % 360) - 180))
ds = ds.sortby("lon")

# 2. Compute annual means to reduce data size before heavy spatial masking
# This converts monthly data into one average map per year
ds_annual = ds.tas.groupby("time.year").mean(dim="time")

# 3. Create a cell-area weight matrix based on latitude (essential for climate data)
weights = np.cos(np.deg2rad(ds_annual.lat))
weights.name = "weights"
ds_annual_weighted = ds_annual.weighted(weights)

# 4. Load country boundaries
world_regions = regionmask.defined_regions.natural_earth_v5_0_0.countries_110
world = world_regions.to_geodataframe()

# 5. Generate the 3D spatial mask grid
mask = regionmask.mask_3D_geopandas(world, ds_annual.lon, ds_annual.lat)

# 6. Extract area-weighted mean values for each country and year
# Resulting shape: (region, year)
country_annual_means = ds_annual_weighted.mean(dim=("lat", "lon"))
country_annual_means = country_annual_means.weighted(mask).mean(
    dim=("lat", "lon")
)

# 7. Convert the xarray DataArray directly into a clean pandas DataFrame
df = country_annual_means.to_dataframe(name="mean_temp").reset_index()

# 8. Map the numeric region indices back to actual country names
df["country"] = df["region"].map(world["names"])

# 9. Clean up and reorder columns
df = df[["year", "country", "mean_temp"]].dropna(subset=["country"])
df = df.sort_values(by=["country", "year"]).reset_index(drop=True)

print(df.shape)
```

```python
# 1. Standardize longitudes to [-180, 180] and sort on the full dataset
ds = ds.assign_coords(lon=(((ds.lon + 180) % 360) - 180))
ds = ds.sortby("lon")

# 2. Compute annual means (Kelvin)
ds_annual = ds.tas.groupby("time.year").mean(dim="time")

# 3. Create cell-area weight matrix based on latitude
weights = np.cos(np.deg2rad(ds_annual.lat))

# 4. Load country boundaries
world_regions = regionmask.defined_regions.natural_earth_v5_0_0.countries_110
world = world_regions.to_geodataframe()

# 5. Generate the 3D spatial mask grid
mask = regionmask.mask_3D_geopandas(world, ds_annual.lon, ds_annual.lat)

# --- FIXED STEPS ---

# 6. Combine grid area weights and country mask into a joint weight matrix
# This zeroes out grid cells outside the country while keeping physical sizing intact
total_weights = mask * weights

# Apply the joint weights and take the mean across space
country_annual_means = ds_annual.weighted(total_weights).mean(dim=("lat", "lon"))

# 7. Convert to DataFrame and change Kelvin to Celsius
df = country_annual_means.to_dataframe(name="mean_temp_kelvin").reset_index()
df["mean_temp"] = df["mean_temp_kelvin"] - 273.15  # Convert to Celsius

# 8. Map numeric region indices back to country names
df["country"] = df["region"].map(world["names"])

# 9. Clean up and reorder columns
df = df[["year", "country", "mean_temp"]].dropna(subset=["country"])
df = df.sort_values(by=["country", "year"]).reset_index(drop=True)

print(df.head())
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import regionmask
import xarray as xr

# 1. Store your sliced data and drop the single time dimension
single_month_temp = ds.tas.sel(time="1950-01").squeeze()

# 2. Shift longitudes to [-180, 180] to match geographic standards
single_month_temp = single_month_temp.assign_coords(
    lon=(((single_month_temp.lon + 180) % 360) - 180)
)
single_month_temp = single_month_temp.sortby("lon")

# 3. Load country boundaries directly from regionmask internal library
world_regions = regionmask.defined_regions.natural_earth_v5_0_0.countries_110
world = world_regions.to_geodataframe()

# 4. Generate a spatial mask grid matching your coordinates
mask = regionmask.mask_3D_geopandas(
    world, single_month_temp.lon, single_month_temp.lat
)

# 5. Extract area-weighted mean values for each country region
country_averages = single_month_temp.weighted(mask).mean(dim=("lat", "lon"))

# 6. FIXED: Safe mapping using index alignment instead of raw arrays
# Converts the DataArray to a Series where the index matches the 'world' dataframe rows
averages_series = country_averages.to_series()
world["jan_1950_temp"] = world.index.map(averages_series)

# 7. Generate and display the Choropleth plot
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(
    column="jan_1950_temp",
    cmap="YlOrRd",
    legend=True,
    legend_kwds={
        "label": "January 1950 Temperature Data",
        "orientation": "horizontal",
    },
    missing_kwds={"color": "lightgrey"},  # Missing small islands colored grey
    ax=ax,
)

ax.set_title(
    "Global Country Mean Temperatures — January 1950", fontsize=16, pad=20
)
ax.axis("off")
plt.show()
```

```python
import numpy as np
import xarray as xr

# 1. Ensure longitudes are sorted and aligned (using your month slice)
single_month_temp = ds.tas.sel(time="1950-01").squeeze()

# 2. Extract the absolute latitude to handle North and South symmetrically
abs_lat = np.abs(single_month_temp.lat)

# 3. Create logical masks for each geographic zone
is_tropical = abs_lat <= 23.5
is_subtropical = (abs_lat > 23.5) & (abs_lat <= 35.0)
is_temperate = (abs_lat > 35.0) & (abs_lat <= 66.5)
is_polar = abs_lat > 66.5

# 4. Filter the temperature array using the masks
# .where() drops all cells outside the condition, leaving only that zone
tropical_temp = single_month_temp.where(is_tropical)
subtropical_temp = single_month_temp.where(is_subtropical)
temperate_temp = single_month_temp.where(is_temperate)
polar_temp = single_month_temp.where(is_polar)

# 5. Calculate area-weighted averages for each zone
# This accounts for grid cells shrinking near the poles
weights = np.cos(np.deg2rad(single_month_temp.lat))

zonal_means = {
    "Tropical": tropical_temp.weighted(weights).mean(dim=("lat", "lon")).item(),
    "Subtropical": subtropical_temp.weighted(weights)
    .mean(dim=("lat", "lon"))
    .item(),
    "Temperate": temperate_temp.weighted(weights)
    .mean(dim=("lat", "lon"))
    .item(),
    "Polar": polar_temp.weighted(weights).mean(dim=("lat", "lon")).item(),
}

# 6. Display the final calculations
for zone, val in zonal_means.items():
    print(f"{zone} Zone Mean Temperature: {val:.2f}")
```

```python
import matplotlib.pyplot as plt

# Assign integer values to each zone to create a categorical map grid
zone_grid = xr.zeros_like(single_month_temp)
zone_grid = zone_grid.where(~is_tropical, 1)  # Tropics = 1
zone_grid = zone_grid.where(~is_subtropical, 2)  # Subtropics = 2
zone_grid = zone_grid.where(~is_temperate, 3)  # Temperate = 3
zone_grid = zone_grid.where(~is_polar, 4)  # Polar = 4

# Map the layout grid
zone_grid.plot(cmap="Accent", cbar_kwargs={"ticks": [1, 2, 3, 4]})
plt.title("Geographic Latitudinal Bands")
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import regionmask

# 1. Load country boundaries from regionmask
world_regions = regionmask.defined_regions.natural_earth_v5_0_0.countries_110
world = world_regions.to_geodataframe()

# 2. Extract the absolute latitude of each country's geographic center
# We use the absolute value to handle Northern and Southern hemispheres equally
country_latitudes = np.abs(world.geometry.centroid.y)

# 3. Define a classification function based on latitudinal thresholds
def classify_zone(lat):
    if lat <= 23.5:
        return "Tropical"
    elif lat <= 35.0:
        return "Subtropical"
    elif lat <= 66.5:
        return "Temperate"
    else:
        return "Polar"


# 4. Map the classification onto the countries DataFrame
world["Climate_Zone"] = country_latitudes.apply(classify_zone)

# 5. Generate the Choropleth map
fig, ax = plt.subplots(1, 1, figsize=(15, 8))

# Define a discrete, distinct color map for the categories
# Options: 'Set2', 'Accent', 'Pastel1', 'Paired'
world.plot(
    column="Climate_Zone",
    cmap="Set2",
    legend=True,
    legend_kwds={
        "title": "Climate Zones",
        "loc": "lower left",
        "bbox_to_anchor": (0.02, 0.1),
    },
    edgecolor="white",  # Adds clean white borders between countries
    linewidth=0.5,
    ax=ax,
)

# Clean up visual elements
ax.set_title(
    "Global Countries Grouped by Latitudinal Climate Zone",
    fontsize=16,
    pad=20,
)
ax.axis("off")
plt.show()
```

```python
df_area = df.query("variable_id == 'areacella' & source_id == 'CESM2'")
ds_area = xr.open_zarr(gcs.get_mapper(df_area.zstore.values[0]), consolidated=True)

df_pr = df.query("variable_id == 'pr' & source_id == 'CESM2' & experiment_id == 'historical' & member_id == 'r1i1p1f1'")
ds_pr = xr.open_zarr(gcs.get_mapper(df_pr.zstore.values[0]), consolidated=True)

total_area = ds_area.areacella.sum(dim=['lon', 'lat'])
pr_timeseries = (ds_pr.pr * ds_area.areacella).sum(dim=['lon', 'lat']) / total_area * 86400
pr_seasonal = pr_timeseries.groupby('time.month').mean()
pr_seasonal.load()

def band_mean(lat_min, lat_max):
    area = ds_area.areacella.sel(lat=slice(lat_min, lat_max))
    pr   = ds_pr.pr.sel(lat=slice(lat_min, lat_max)) * 86400
    return (pr * area).sum(['lat','lon']) / area.sum()
tropics   = band_mean(-23.5, 23.5).groupby('time.month').mean().load()
midlat_nh = band_mean(40, 60).groupby('time.month').mean().load()
```

```python
tropics.plot(label='Tropics (23.5°S–23.5°N)')
midlat_nh.plot(label='NH Mid-lats (40°–60°N)')
plt.legend()
plt.title('Seasonal Cycle of Precipitation')
plt.ylabel('mm/day')
plt.xticks(range(1,13), range(1,13))
```

```python
df_pointer = df.query("activity_id=='CMIP' & table_id == 'Amon' & variable_id == 'tas' & experiment_id == 'historical' & source_id == 'CESM2' & member_id == 'r1i1p1f1'")
ds_proj = xr.open_zarr(gcs.get_mapper(df_pointer.zstore.values[0]), consolidated=True)
regions = {
    "tropical": (-23.5, 23.5),
    "subtropical": (23.5, 40),
    "temperate": (40, 60),
    "polar": (60, 90)
}
#region classifications from https://content.meteoblue.com/en/research-education/educational-resources/meteoscool/general-climate-zones
#also the baseline is the modern nasa baseline 
items = []

for region, (low, high) in regions.items():
    mask = (np.abs(ds.lat) >= low) & (np.abs(ds.lat) < high)
    spec = ds.tas.where(mask, drop=True)
    spec_mean = spec.mean(dim=['lat', 'lon'])
    yearly = spec_mean.groupby('time.year').mean()
    baseline = spec_mean.sel(time=slice("1951", "1980")).mean(dim="time")
    spec_anom = spec_mean - baseline
    spec_anom_yearly = spec_anom.groupby("time.year").mean()
    for year in yearly['year'].values:
        items.append([
            region,
            int(year),
            float(yearly.sel(year=year)),
            float(spec_anom_yearly.sel(year=year))
        ])
clean_df = pd.DataFrame(items, columns = ['region', 'year', 'mean_temp', 'anomaly'])

    
```

```python
##Making the cleaned df for this

regions = {
    "tropical":    (-23.5, 23.5),
    "subtropical": (23.5,  40),
    "temperate":   (40,    60),
    "polar":       (60,    90)
}

df_pointer = df.query(
    "table_id == 'Amon' & variable_id == 'tas' & "
    "member_id == 'r1i1p1f1' & source_id == 'CESM2-WACCM'"
)
experiments = ["historical", "ssp126", "ssp245", "ssp585"]

# Compute Base data once
print("Step 1 of 2 · computing baselines from historical 1951–1980...")
baselines = {}
hist_subset = df_pointer[df_pointer["experiment_id"] == "historical"]

for _, row in hist_subset.iterrows():
    ds_hist = xr.open_zarr(gcs.get_mapper(row["zstore"]), consolidated=True)
    tas_hist = ds_hist["tas"]
    for region, (low, high) in regions.items():
        mask = (np.abs(ds_hist['lat']) >= low) & (np.abs(ds_hist['lat']) < high)
        spec = tas_hist.where(mask, drop=True)
        spec_mean = spec.mean(dim=['lat', 'lon'])
        yearly = spec_mean.groupby('time.year').mean()
        baselines[region] = float(
            yearly.sel(year=slice(1951, 1980)).mean(skipna=True)
        )

for r, b in baselines.items():
    print(f"   {r:>11}: {b:.2f} K")

# Build Full DF
print("\nStep 2 of 2 · building dataframe across all experiments...")
rows = []
for exp in experiments:
    print(f"   processing {exp}...")
    subset = df_pointer[df_pointer["experiment_id"] == exp]
    for _, row in subset.iterrows():
        ds = xr.open_zarr(gcs.get_mapper(row["zstore"]), consolidated=True)
        tas = ds["tas"]
        for region, (low, high) in regions.items():
            mask = (np.abs(ds['lat']) >= low) & (np.abs(ds['lat']) < high)
            spec = tas.where(mask, drop=True)
            spec_mean = spec.mean(dim=['lat', 'lon'])
            yearly = spec_mean.groupby('time.year').mean()

            
            anomaly = yearly - baselines[region]

            for y in yearly['year'].values:
                rows.append([
                    region, exp, int(y),
                    float(yearly.sel(year=y)),
                    float(anomaly.sel(year=y)),
                ])

clean_df = pd.DataFrame(
    rows, columns=['region', 'experiment', 'year', 'mean_temp', 'anomaly']
)



clean_df.head(10)
```
